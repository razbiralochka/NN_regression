Будем делать регрессию в библиотеке TensorFlow, которая позволит нам создать полносвязную нейросеть

```python
import tensorflow as tf
import numpy as np


```

Тут мы импортируем наш Датасет для обучения а он лежит в файле Train.txt
И удалим битые строки

```python
DataSet = np.genfromtxt('Train.txt')
indexList = [np.any(i) for i in np.isnan(DataSet)]
DataSet = np.delete(DataSet, indexList, axis=0)
print(DataSet)
```

Итак вот есть матрица а нам нужна завсисимость расхода бензина от параметоров ТС.
Для этого будем слайсить массив

```python
mpg = DataSet[:,0]
print(mpg)
```

Готово! Тоже самое для параметоров машины)

```python
car = DataSet[:,1:8]
print(car)
```

Всё! Данные разделены теперь нам надо составить модель

```python
model = tf.keras.models.Sequential([
  tf.keras.layers.Dense(64, activation='relu',input_dim = 7),
  tf.keras.layers.Dense(128, activation='relu'),
  tf.keras.layers.Dense(1)
])
```

Итак разбираем что тут. Первый слой состоит из 100 нейронов и 7 входов([  8.  307.  130.  ...  12.   70.    1. ]). У каждого на выходе линейная функция

Второй слой 300 нейронов и функция на выходе ReLU
А это уже ломаная прямуля блин)))


Последний слой содежит один нейрон (потому что и выход один: миль на галлон)
А сама нейросеть вот она:

```python
model.summary()
```

Теперь укажем как нейросеть будет обучаться. А обучаться она будет методом адаптивного градиентного спуска минимизируя абсолютную ошибку:

```python
model.compile(loss='mean_absolute_error',optimizer=tf.keras.optimizers.Adam(0.0001))
```

Теперь скормим данные нейросети и сделаем 1000 итерации обучения:

```python
model.fit(car, mpg, epochs=2000,batch_size = 100)
```

Окей, она типо обучена...теперь загружаем Датасет тестовый

```python
TestDataSet=np.genfromtxt('Test.txt')
indexList = [np.any(i) for i in np.isnan(TestDataSet)]
TestDataSet = np.delete(TestDataSet, indexList, axis=0)
print(TestDataSet)
```

```python
test_cars = TestDataSet[:,1:8]
print(test_cars)
```

Загоняем всё это в нейронку и сравниваем ответы

```python
results = model.predict(test_cars)
```

```python
err= TestDataSet[:,0]-results[0,:]
print(err)
```

```python
r_err = abs(err)/results[0,:]
print(r_err)
```

Нейросеть выдала данные и погреностью 1-70%
